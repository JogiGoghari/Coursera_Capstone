```python
# this task will only performed when you have your csv file uplodeted in this tree 
# you can see "Toronto.TASK_1_df.csv"  uploded or created by another notebook in the tree 
task1_data_csv = "Toronto.TASK_1_df.csv"
df1 = pd.read_csv(task1_data_csv).set_index("Postal Code")
df1
```

```python
df1.rename_axis("Postal Code", axis='index', inplace=True)
df1.head()
```

## Now we are connecting above dataframe with Latitude & Longitude , where we have this given link for use of it   "http://cocl.us/Geospatial_data" 

```python
filename = "http://cocl.us/Geospatial_data"
```

```python
df2 = pd.read_csv(filename)
df2
```

#### So, here
#### df1 = is a data of toronto with postal code and other details
#### df2 = same postal codes with their Latitude & Longitude  

### Left merging "df1" and "df2" (dataframes) to add Latitude and Longitude values to df1 dataframe and name the merged dataframe as as "df3"

```python
df3 = pd.merge(df1, df2, on ='Postal Code', how ='left')
df3
```

```python
len(df3["Latitude"])
```

```python
# to convert this file in to csv verstion 
df3.to_csv('Toronto.TASK_2_df.csv',index=False)
```
